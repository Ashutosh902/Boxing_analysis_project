#### After proper analysis of data the observation was that data was containing too many nulls in most of the columns and below are some of the methods being adopted to remove them:
1. Age column had nulls in it which was either for both the players or either one of them.
2. Based on this observation I took the mean of the ages for player A and B seperately
3. Based on that mean I took the difference between the mean of both the players(i.e player and opponent) 
4. After this did a analysis of the players based on their age to find out the correlation and reached to a conclusion that the players who have less age are having probability of winning.
5. After this observation applied the reverse analysis for the data i.e putting the difference of mean in the place of the players whose date is null by replacing it with (mean difference + players age if he has lost the match and vice versa)

Note : Simillar trend I followed to fill the missing weight and height too.

### Handling the extreme values:
1. As the data was for Boxing match i did a small study about the criteria and age boundaries being followed by most of the organizations which conducts the boxing matches.
2. During my study i found that the age limit followed by most of those organizations were falling in the range of 13 to 40 years and also their weight and height will fall in the category of more than 80 pound and between 50 to 218 cm respectively.

```python
box_full_data = pd.read_csv('box_data_20.csv')
```

```python
for i in box_full_data[box_full_data['age_A'] <13].index:
    box_full_data.loc[i, 'age_A'] = 13
for i in box_full_data[box_full_data['age_B'] <13].index:
    box_full_data.loc[i, 'age_B'] = 13
for i in box_full_data[box_full_data['age_A'] >40].index:
    box_full_data.loc[i, 'age_A'] = 40
for i in box_full_data[box_full_data['age_B'] >40].index:
    box_full_data.loc[i, 'age_B'] = 40
```

```python
box_full_data.drop('Unnamed: 0', axis =1, inplace= True)
```

```python
for i in box_full_data[box_full_data['weight_A'] <80].index:
    box_full_data.loc[i, 'weight_A'] = 80
for i in box_full_data[box_full_data['weight_B'] <80].index:
    box_full_data.loc[i, 'weight_B'] = 80
```

```python
for i in box_full_data[box_full_data['height_A'] <50].index:
    box_full_data.loc[i, 'height_A'] = 50
for i in box_full_data[box_full_data['height_B'] <50].index:
    box_full_data.loc[i, 'height_B'] = 50
for i in box_full_data[box_full_data['height_A'] >218].index:
    box_full_data.loc[i, 'height_A'] = 218
for i in box_full_data[box_full_data['height_B'] >218].index:
    box_full_data.loc[i, 'height_B'] = 218
```
